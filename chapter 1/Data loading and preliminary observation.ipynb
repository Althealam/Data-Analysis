逐块读取（Chunking）是指将大型数据文件分割成较小的块（或批次），然后逐个处理这些块<br>

逐块处理的优点：<br>
1.内存效率：当处理大型数据文件时，一次性将整个文件加载到内存可能会导致内存不足的问题。通过逐块读取，你可以将数据分成较小的块，并逐个处理，从而减少对内存的要求。<br>

2.增量处理：逐块读取允许你逐个处理数据块，并对每个块执行特定的操作。这对于实时或增量处理数据非常有用，因为你可以在处理一个块后立即获取结果，而无需等待整个数据集加载和处理完成。<br>

3.迭代处理：通过逐块读取，你可以使用循环来处理每个块。这意味着你可以在处理一个块后进行一些操作，然后再处理下一个块，以此类推。这种迭代方式非常灵活，允许你在处理数据时逐步构建结果或执行复杂的数据转换。<br>

#### reset_index函数的作用<br>
reset_index() 是 Pandas 中的一个函数，用于重置数据帧（DataFrame）的索引。<br>
它的作用是将当前的索引重新设置为默认的整数索引，并将原来的索引作为一个新的列添加到数据帧中。<br>
默认情况下，数据帧的索引是整数序列（0、1、2、...），如果在处理数据时对索引进行了修改或者使用了某些操作导致索引不再是连续的整数序列，那么可以使用 reset_index() 函数将索引重置为默认的整数序列

```python
#重新设置索引
midage = midage.reset_index(drop=True)
#利用drop参数丢弃原来的索引，重新设置索引
midage.head(3)
midage.loc[[100],['Pclass','Sex']]
```

#### loc函数的作用<br>
通过行标签和列标签来访问数据帧（DataFrame）中的特定行和列，或者对它们进行切片和筛选

#### iloc与loc的异同

- **索引方式**：
   - `iloc` 使用基于位置的索引，即通过整数位置来定位行和列。
   - `loc` 使用基于标签的索引，即通过行标签和列标签来定位行和列。

- **使用方式**：
   - `iloc` 的使用方式类似于Python中的列表索引，它通过指定整数位置来选择数据。
   - `loc` 的使用方式更接近于字典的键值对，它通过指定行标签和列标签来选择数据。

- **切片操作**：
   - `iloc` 使用的是半开区间（左闭右开）的切片操作。
   - `loc` 使用的是闭区间的切片操作。

- **性能**：
   - `iloc` 在使用整数位置索引时通常会比 `loc` 快。
   - `loc` 在使用标签索引时可能稍慢，因为它需要进行额外的标签匹配操作。
